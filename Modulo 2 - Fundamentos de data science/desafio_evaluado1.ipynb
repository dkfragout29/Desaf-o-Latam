```python
primer_array=np.linspace(1, 50, 50)

segundo_array=np.linspace(50, 150, 101)
```

```python
for i in segundo_array:
    if i%2==0:
        print(f'El número {i} es divisible 2')
    else:
        i%3==0
        print(f'El número {i} es divisible por 3')
```

```python
for i in segundo_array:
    if i%2==0 and i%3==0:
        print(f'El número {i} es divisible 2 y por 3')
```
